# Prepping Data Challenge - Week 02

## Import Libraries

```python
import numpy as np
import pandas as pd
```

## Import Data

```python
flow_card_holders = pd.read_csv('D:01_Projekt_Portfolio/data_prepping_challenges/flow_card_holders.csv')
flow_card_haters = pd.read_csv('D:01_Projekt_Portfolio/data_prepping_challenges/flow_card_haters.csv')
```

## Challenges

### 1. UNION the data back together

We will do this with `pd.concat`

```python
data = pd.concat([flow_card_holders.copy(), flow_card_haters.copy()]).reset_index(drop=True)
```

### 2. Convert `Date` to Quarter

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data['Quarter'] = data['Date'].dt.quarter
```

```python
data
```

### 3. Group by Flow Card, Quarter and Class

```python
minimum_price = data.groupby(['Quarter', 'Flow Card?', 'Class']).Price.min().unstack()
median_price = data.groupby(['Quarter', 'Flow Card?', 'Class']).Price.median().unstack()
maximum_price = data.groupby(['Quarter', 'Flow Card?', 'Class']).Price.max().unstack()
```

#### Add a column which shows Price type

```python
minimum_price['Price Type'] = 'minimum'
median_price['Price Type'] = 'median'
maximum_price['Price Type'] = 'maximum'
```

#### Concat the views back together

```python
quarter_report = pd.concat([minimum_price, median_price, maximum_price])
```

### 4. Change Column Names to correct class

Map Columns

```python
quarter_report.columns = quarter_report.columns.map({'First Class' : 'Economy', 'Business Class' : 'Premium', 'Premium Economy' : 'Business', 'Economy' : 'First', 'Price Type' : 'Price Type'})
```

Sort Columns

```python
quarter_report = quarter_report[['Economy', 'Premium', 'Business', 'First', 'Price Type']]
```
