# Prepping Data Challenge - Week 16
**Budget vs. Actuals**

## Import Libraries

```python
import numpy as np
import pandas as pd
```

## Import Data

```python
budget = pd.read_csv('D:01_Projekt_Portfolio/data_prepping_challenges/inputs/PD_2024_Wk_16_Budget.csv')
actual = pd.read_csv('D:01_Projekt_Portfolio/data_prepping_challenges/inputs/PD_2024_Wk_16_Actual.csv')
```

```python
actual
```

```python
budget = budget.dropna()
budget
```

## Challenges

### Budget Column formatting

```python
budget['Budget'] = budget['Budget'].apply(lambda x: int(x.replace('K', '000').replace(',','')))
```

```python
actual['Actual'] = actual['Actual'].apply(lambda x: float(x.replace(',', '')))
```

### Join datasets

```python
comparison = budget.merge(actual, on='2022')

# Drop the 'Notes' column
comparison = comparison[['2022', 'Budget', 'Actual']]

# Change the column names
comparison.columns = ['Category', 'Forecasted Spending', 'Actual Spending']

# Calculate the difference
comparison['Difference'] = comparison['Actual Spending'] - comparison['Forecasted Spending']
comparison['Difference'] = comparison['Difference'].astype(int)
```
