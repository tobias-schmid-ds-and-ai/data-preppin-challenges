## Import Libraries

```python
import numpy as np
import pandas as pd
```

## Import Data

No real imports this time. The goal of this challenge is to check how months would switch, when we use 13 months instead of 12

```python
date_data = pd.DataFrame({'Date' : pd.date_range(np.datetime64('2024-01-01'), np.datetime64('2024-12-31'))})
```

### Create new 13 / 14 month system

It seems like the simplest approach would be to just make a list of Dates

```python
new_dates = [f'2024-{month}-{day}' for month in range(1,14) for day in range(1,29)] + ['2024-14-01', '2024-14-02']
        
```

```python
date_data['New Dates'] = new_dates
```

### Filter for the days where the day of the month changes

```python
changing_dates = date_data[date_data['Date'].dt.month != date_data['New Dates'].apply(lambda x: int(x.split('-')[1]))]
```
