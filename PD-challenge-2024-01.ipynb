## Tasks

### 1. Split the flight details field

Target variables:
- `Date`
- `Flight Number`
- `From`
- `To`
- `Class`
- `Price`

#### First idea: use apply method on strings in `Flight Details`

```python
def split_flight_details(flight_details):
    # Split Data into list of variables
    list_of_variables = str(flight_details).split('//')
    
    # To be more prudent, we will assume missing values
    try:
        # Date
        date = np.datetime64(list_of_variables[0])
    except:
        date = np.nan

    try:
        # Flight Number
        flight_number = list_of_variables[1]
    except:
        flight_number = np.nan

    try:
        # From and To
        from_to = list_of_variables[2].split('-')
        from_ = from_to[0]
        to = from_to[1]

    except:
        from_ = np.nan
        to = np.nan

    try:
        # Class
        class_ = list_of_variables[3]

    except:
        class_ = np.nan

    try:
        # Price
        price = float(list_of_variables[4])
    except:
        price = np.nan

    return pd.Series({'Date' : date, 'Flight Number' : flight_number, 'From' : from_, 'To' : to, 'Class' : class_, 'Price' : price})
```

Make the columns

```python
prepared_data[['Date', 'Flight Number', 'From', 'To', 'Class', 'Price']] = data['Flight Details'].apply(split_flight_details)
```

Drop the `Flight Details` column

```python
prepared_data = prepared_data.drop(['Flight Details'], axis=1)
```

### Change Card holder field to Yes / No 

```python
prepared_data['Flow Card?'] = prepared_data['Flow Card?'].map({1 : 'Yes', 0 : 'No'})
```

### Split the data set into two, based on `Flow Card?` value

```python
flow_card_holders = prepared_data[prepared_data['Flow Card?'] == 'Yes']
flow_card_haters = prepared_data[prepared_data['Flow Card?'] == 'No']
```
