# Prepping Data Challenge - Week 07

Valentines Day Challenge

## Import Libraries

```python
import numpy as np
import pandas as pd
```

## Import Data

```python
couples = pd.read_csv('D:01_Projekt_Portfolio/data_prepping_challenges/inputs/PD_2024_Wk_7_Couples.csv')
gifts = pd.read_csv('D:01_Projekt_Portfolio/data_prepping_challenges/inputs/PD_2024_Wk_7_Gifts.csv')
```

## Challenges

**Goal:** Find the fitting present for each couple on valentines Day 2024

#### Set fixed valentines Day

```python
valentines_day = np.datetime64('2024-02-14')
```

### Change Data Types

```python
# Relationship start as a Datetime object
couples['Relationship Start'] = pd.to_datetime(couples['Relationship Start'])
```

```python
# Year as integer
gifts['Year'] = gifts.Year.apply(lambda x: int(x[:-2]))
```

### Obtain milestone

```python
couples['Full Years in Relationship'] = (valentines_day - couples['Relationship Start']) // pd.Timedelta(days=365)
```

```python
couples['Milestone'] = couples['Full Years in Relationship'].apply(lambda x: x if x <=10 else x - x % 5)
```

### Merge Datasets

```python
gifts_for_couples = pd.merge(couples, gifts, how='left', left_on='Milestone', right_on='Year')[['Couple', 'Relationship Start', 'Milestone', 'Gift']]
```

## Save Dateset

```python
gifts_for_couples.to_csv('D:01_Projekt_Portfolio/data_prepping_challenges/outputs/07_couples_gifts.csv')
```
