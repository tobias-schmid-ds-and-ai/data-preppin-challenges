## Challenges

### Locate which item is in which store

```python
recalled_items['Recalled'] = True
stock_with_recall = stock_per_store.merge(recalled_items, on=['Category', 'Product ID', 'Unit Price'], how='left')
```

### Calculate how much money was lost

```python
stock_with_recall['Loss'] = np.where(stock_with_recall['Recalled'] == True, stock_with_recall['Unit Price'] * stock_with_recall['Quantity'], np.nan)
```

### Money lost by category

```python
loss_by_category = stock_with_recall.groupby('Category')['Loss'].sum().reset_index()
loss_by_category
```

```python
loss_by_category.to_csv('D:01_Projekt_Portfolio/data_prepping_challenges/outputs/14_loss_by_category.csv')
```

### Loss by store

```python
loss_by_store = stock_with_recall.groupby('Store')['Loss'].sum().reset_index()
```

```python
loss_by_store['Priority'] = np.where(loss_by_store['Loss'] >= 5000, 'High Priority', 'Low Priority')
```
