### Sales Date to datetime

```python
dataframes = [data_2020, data_2021, data_2022, data_2023, data_2024]
```

```python
for df in dataframes:
    df['Sales Date'] = pd.to_datetime(df['Sales Date'], format='%d/%m/%Y')
```

### Add an 'Easter Week' column

```python
# Check if all DataFrames have the same number of weeks and check if each first sales date is a monday
for df in dataframes:
    print(df['Sales Date'].dt.isocalendar().week.nunique())
```

All DataFrames have the same number of weeks

#### Add an Easter Sale Week Number for each Year

```python
for df in dataframes:
    df['Easter Sale Week Number'] = df['Sales Date'].dt.isocalendar().week - df['Sales Date'].dt.isocalendar().week.min() + 1
```

#### Add weekday name and number

```python
for df in dataframes:
    df['Weekday'] = df['Sales Date'].dt.strftime('%a')      # Abbreviation of the weekday
    df['Weekday Order'] = df['Sales Date'].dt.weekday + 1   # Number for each weekday, starting at 1 = Monday
```

### Concatenate DataFrames

```python
all_sales = pd.concat(dataframes).reset_index(drop=True)
```

### Calculate sum of Sales per Day and product

A simple multiplication of price times quantity

```python
all_sales['Sales'] = all_sales['Price'] * all_sales['Quantity Sold']
```
