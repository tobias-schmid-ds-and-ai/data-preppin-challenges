## Import Libraries

```python
import numpy as np
import pandas as pd
```

## Import Data

#### Output Data From Week 1

```python
flow_card_haters = pd.read_csv('D:01_Projekt_Portfolio/data_prepping_challenges/flow_card_haters.csv')
flow_card_holders = pd.read_csv('D:01_Projekt_Portfolio/data_prepping_challenges/flow_card_holders.csv')
```

```python
sales_data = pd.concat([flow_card_haters.copy(), flow_card_holders.copy()]).reset_index(drop=True) 
```

#### New Quarterly Data

```python
q1_data = pd.read_csv('D:01_Projekt_Portfolio/data_prepping_challenges/PD_2024_Wk_3_Input_Q1.csv')
q2_data = pd.read_csv('D:01_Projekt_Portfolio/data_prepping_challenges/PD_2024_Wk_3_Input_Q2.csv')
q3_data = pd.read_csv('D:01_Projekt_Portfolio/data_prepping_challenges/PD_2024_Wk_3_Input_Q3.csv')
q4_data = pd.read_csv('D:01_Projekt_Portfolio/data_prepping_challenges/PD_2024_Wk_3_Input_Q4.csv')
```

```python
target_data = pd.concat([q1_data.copy(), q2_data.copy(), q3_data.copy(), q4_data.copy()]).reset_index(drop=True)
```

## Challenges

### 1. Correct the classes of flights

```python
sales_data['Class'] = sales_data['Class'].map({'Economy' : 'First Class', 'First Class' : 'Economy', 'Premium Economy' : 'Business Class', 'Business Class' : 'Premium Economy'})
```

### 2. Adjust the Flight classes to the first letter pattern

```python
# Hard to read but elegant solution: Take the first letter of every word
sales_data['Class'] = sales_data['Class'].apply(lambda x: "".join([w[0] for w in x.split(' ')]))
```

```python
sales_data
```

### 3. Create a month Column on which to join

```python
sales_data['Date'] = pd.to_datetime(sales_data['Date'])
```

```python
sales_data['Month'] = sales_data['Date'].dt.month
```

### 4. Total sales by class and month

```python
sales_by_class_and_month = sales_data.groupby(['Month', 'Class'])['Price'].sum().reset_index()
```

```python
sales_vs_target = pd.merge(sales_by_class_and_month, target_data, on=['Month', 'Class'])
sales_vs_target.columns = ['Month', 'Class', 'Sales', 'Target']
```

```python
sales_vs_target['Difference to Target'] = sales_vs_target['Sales'] - sales_vs_target['Target']
```
