# Prepping Data Challenge - Week 09

## Import Libraries

```python
import numpy as np
import pandas as pd
```

## Import Data

```python
actions = pd.read_csv('D:01_Projekt_Portfolio/data_prepping_challenges/inputs/PD_2024_Wk_9_Customer_Actions.csv')
flight_details = pd.read_csv('D:01_Projekt_Portfolio/data_prepping_challenges/inputs/PD_2024_Wk_9_Flight_Details.csv')
```

### Cast datetime columns

```python
actions['Flight Date'] = pd.to_datetime(actions['Flight Date'], format='%d/%m/%Y')
actions['Date'] = pd.to_datetime(actions['Date'], format='%d/%m/%Y')
flight_details['Flight Date'] = pd.to_datetime(flight_details['Flight Date'], format='%d/%m/%Y')
```

## Challenges

```python
last_actions = actions.sort_values(by='Date').drop_duplicates('Customer ID', keep='last')[actions['Action'] != 'Cancelled']
```

```python
merged_data_set = pd.merge(last_actions, flight_details, on=['Flight Number', 'Flight Date', 'Class'], how='left')
```

```python
merged_data_set = merged_data_set.sort_values(by= ['Flight Number', 'Class','Date'])
```

```python
section = merged_data_set['Flight Number'] + ' ' + merged_data_set['Class']
```

```python
section_dummies = pd.get_dummies(section)
```

```python
running_booking = section_dummies.cumsum()
```

```python
running_booking
```

```python
def get_booking(customer, running_booking):
    section = customer['Flight Number'] + ' ' + customer['Class']
    return running_booking.loc[customer.name, section]
```

```python
merged_data_set['Total Seats booked over time'] = merged_data_set.apply(lambda x: get_booking(x, running_booking), axis=1)
```

```python
merged_data_set['Percentage of Seats booked'] = merged_data_set['Total Seats booked over time'] / merged_data_set['Capacity']
```

## Save Data

```python
merged_data_set.to_csv('D:01_Projekt_Portfolio/data_prepping_challenges/outputs/09_seat_bookings.csv')
```
